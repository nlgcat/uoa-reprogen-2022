```python
from csv import DictReader
import pandas as pd
import numpy as np

import pprint
pp = pprint.PrettyPrinter(indent=2)

test_set_ids = {'T'+str(i).zfill(3)+'.txt':'S'+str(i+60).zfill(3)+'.txt' for i in range(1,31)}

def format_exp_id(exp_id):
  return {
    '00_Existing':  'A',
    '01_To_60':     'B',
    '02_To_90':     'C',
  }[exp_id]


def format_sys_id(sys_id):
  return {
    'wiseman_frototest_conditional_copy_beam5_gens.txt':  'conditional_copy',
    'puduppully_aaai_19_rotowire_test.txt':               'document_plan',
    'rebuffel_test_gen_not_paper.txt':                    'hierarchical_encoder',
  }[sys_id]

def format_doc_id(doc_id):
  d = doc_id
  if doc_id in test_set_ids:
    d = test_set_ids[doc_id]

  return d.replace('.txt','')

with open('docs.csv', 'r') as fh:
  rows = list(DictReader(fh))
  doc_experiments = {format_doc_id(row['DOC_ID']):format_exp_id(row['EXP_ID']) for row in rows}
  doc_systems = {format_doc_id(row['DOC_ID']):format_sys_id(row['SYS_ID']) for row in rows}


results = []
with open('errors.csv', 'r') as fh:
  reader = DictReader(fh)

  for row in reader:
    doc_id = format_doc_id(row['TEXT_ID'])
    sys_id = doc_systems[doc_id]
    exp_id = doc_experiments[doc_id]
    error_type = row['TYPE']
    results.append([doc_id,sys_id,exp_id,error_type])

error_types = set()

df = pd.DataFrame(results, columns=['DOC_ID','SYS_ID', 'EXP_ID', 'TYPE'])

print(df)
print(df.agg('std'))

exps = df.groupby(['EXP_ID','SYS_ID'])

experiments_sizes = {'A': 21, 'B':39, 'C':30}

final_results = []
for row in exps.count().iterrows():
  exp_id = row[0][0]
  sys_id = row[0][1]
#   error_type = row[0][2]
  error_type = None
  error_types.add(error_type)
  v = (row[1]['DOC_ID'] / experiments_sizes[exp_id]).round(2)
  final_results.append((exp_id,sys_id,error_type,v))

print(error_types)
pp.pprint(final_results)
```

```python
import matplotlib.pyplot as plt

mean_x = list(range(0,50))

system = 'hierarchical_encoder'

experiments = ['A', 'B', 'C']
systems = ['conditional_copy', 'document_plan', 'hierarchical_encoder']
#categories = ['NAME', 'NUMBER', 'WORD', 'CONTEXT', 'NOT_CHECKABLE', 'OTHER']
categories = [None]

h = {}

for cat in categories:
    h[cat] = {}
    for system in systems:
        h[cat][system] = {}
        for exp in experiments:
            h[cat][system][exp] = 0.0

for r in final_results:
    exp_id = r[0]
    sys_id = r[1]
    error_type = r[2]
    v = r[3]

        
    h[error_type][sys_id][exp_id] = v
    
pp.pprint(h)
    
for error_type, x in h.items():
    plt.figure()
    
    
    for sys_id, data in x.items():
        keys = [k for k in data.keys()]
        vals = [v for v in data.values()]
        plt.plot(keys, vals, label=sys_id)
        plt.title(error_type)
        plt.xlabel('experiment')
        plt.ylabel('mean')
    plt.ylim(ymin=0)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
```

```python
doc_counts = {sys_id:{exp_id:0 for exp_id in experiments} for sys_id in systems}

full_count = {exp_id:0 for exp_id in experiments}

categories = ['NAME', 'NUMBER', 'WORD', 'CONTEXT', 'NOT_CHECKABLE', 'OTHER']
full_type_count = {exp_id:{cat_id:0 for cat_id in categories} for exp_id in experiments}

for result in results:
    doc_id = result[0]
    sys_id = result[1]
    exp_id = result[2]
    category = result[3]
    doc_counts[sys_id][exp_id] += (1 / (experiments_sizes[exp_id] / 3))
    full_count[exp_id] += (1 / experiments_sizes[exp_id])
    
pp.pprint(full_count)
pp.pprint(full_type_count)
pp.pprint(doc_counts)
pp.pprint(results)
```
